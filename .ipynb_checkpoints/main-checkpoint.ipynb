## Data Processing

```python
import pandas as pd

df = pd.read_excel('./results/editeurs_de_logiciels_from_archimag_with_more_inf_18-11-2024_19h30min.xlsx')
df = df.drop_duplicates(subset=['name'], keep='first')
df.head(2)
```

```python
def contact_process(contact):
    if contact:
        res = str(contact).replace('Nous contacter :\n','').replace('Nous contacter :','').replace('\xa0',' ').strip()
        return res if res != '' else None
    else:
        return contact

def extract_contact_info(contact):
    if contact:
        tele = None
        email = None
        address = None
        web_site_url = None
        
        res = str(contact).split('Téléphone: \n')
        if len(res)>1:
            tele = res[1].split('\n')[0]

        res = str(contact).split('Courriel: \n')
        if len(res)>1:
            email = res[1].split('\n')[0]
            
        res = str(contact).split('Téléphone: \n')
        if len(res)>1:
            address = res[0].replace('\n', ' ')
        else:
            res = str(contact).split('Courriel: \n')
            if len(res)>1:
                address = res[0].replace('\n', ' ')
            else:
                address = contact.replace('\n', ' ')

        res = str(contact).split('\n')
        if len(res)>1:
            web_site_url = res[-1] if 'http' in res[-1] or 'www.' in res[-1] else None
            
        return tele, email, address, web_site_url
        
    else:
        return contact, contact, contact, contact

df['contact'] = df['contact'].apply(contact_process)
#df['tele'] = df['contact'].apply(extract_tele)

df[['tele', 'email', 'address', 'web_site_url']] = df['contact'].apply(lambda x: pd.Series(extract_contact_info(x)))
df.head(2)
```

```python
def commercial_name_process(commercial_contact):
    if commercial_contact:
        return ''.join([commercial.replace('\n',' : ') for commercial in str(commercial_contact).split('Contact Commercial :\n') if commercial.strip() not in['', 'Contact Commercial :']])
    else:
        return commercial_contact
    
df['commercial_name'] = df['commercial_name'].apply(commercial_name_process)
df.head(2)
```

# Extract more information from linkedin ( linkedin url, profiles, size, ...)

```python
from ItemScraping import ItemScraping
from Item import EditeurLogiciels
from ItemStorage import ItemStorage
from LinkedinAutomation import LinkedinAutomation
import time, random

import os
from dotenv import load_dotenv

import copy
```

```python
# Charger les variables d'environnement depuis le fichier .env
load_dotenv('settings.env')

# Accéder aux variables d'environnement
email = os.getenv('LINKEDIN_EMAIL')
pwd = os.getenv('LINKEDIN_PWD')

linkedinAutomation = LinkedinAutomation()
#linkedinAutomation.driver.quit()
linkedinAutomation.get_linkedin_authentication(email = email, pwd = pwd)
```

```python
# itemStorage = ItemStorage(file_path='./results/editeurs_de_logiciels_7_have_between_51_and_1000_employees_25-11-2024.csv')
# dicts = itemStorage.get_list_of_dicts()
# company_linkedin_url = dicts[4].get('linkedin')
```

```python
def difference_dic(dic1, dic2, column_name=''):
    # Extraire les valeurs de 'Nom_de_l_entreprise' de dic2 dans un set pour une recherche rapide
    entreprises_dic2 = {d[column_name] for d in dic2}
    # Retourner les dictionnaires dans dic1 qui ne sont pas dans dic2 (selon 'Nom_de_l_entreprise')
    return [d for d in dic1 if d[column_name] not in entreprises_dic2]
    
# itemStorage = ItemStorage(file_path='./results/editeurs_de_logiciels_9_with_linkedin_about_info_25-11-2024.csv')
# dicts_1 = itemStorage.get_list_of_dicts()
# itemStorage = ItemStorage(file_path='./results/editeurs_de_logiciels_10_with_linkedin_about_info_25-11-2024.csv')
# dicts_2 = itemStorage.get_list_of_dicts()

# dicts = difference_dic(dicts_1, dicts_2, column_name='name')

itemStorage = ItemStorage(file_path='./results/editeurs_de_logiciels_12_have_between_51_and_1000_employees_with_linkedin_about_info_25-11-2024.csv')
dicts = itemStorage.get_list_of_dicts()

print(len(dicts))

# task = 'extract_linkedin_url_from_company_web_site_url'
# task = 'extract_linkedin_url_from_google_based_on_company_name'
# task = 'extract_linkedin_about_info'
# task = 'process_linkedin_about_info_column'
# task = 'extract_profiles'
task = 'extract_founder_and_manager_profiles_based_on_keywords'

for i, dict in enumerate(dicts):
    editeurs = []
    editeur = EditeurLogiciels(dict)

    if task == 'extract_linkedin_about_info':
        result = linkedinAutomation.extract_company_info_from_company_linkedin_profile_url(dict.get('linkedin'))
        if result['status']:
            editeur.linkedin_about_info = result['data']
            print(f'{dict.get('name')} ==> {result['data']}')
        else:
            print(f'{dict.get('name')} ==> {result['data']}')
            editeur.linkedin_about_info = None

    if task == 'process_linkedin_about_info_column':
        if str(dict.get('linkedin_about_info')) != 'nan':
            new_dict = { f'about {key}':value for key, value in linkedinAutomation.str_to_json(dict.get('linkedin_about_info')).items()}
            dict.update(new_dict)
        editeur.init_from_dict(dict)
    #############################################################################################################

    elif task == 'extract_profiles':
        result = linkedinAutomation.extract_profiles_from_company_linkedin_profile_url(dict.get('linkedin'))
        if result['status']:
            editeur.profiles = result['data']
            print(f'{dict.get('name')} ==> {len(result['data'])}')
        else:
            print(f'{dict.get('name')} ==> {result['data']}')
            editeur.profiles = None

    elif task == 'extract_linkedin_url_from_company_web_site_url':
        result = linkedinAutomation.get_company_linkedin_url_from_company_web_site_url(dict.get('web_site_url'))
        if result['status']:
            editeur.linkedin = result['data']
            print(f'{dict.get('name')} ==> {result['data']}')
        else:
            print(f'{dict.get('name')} ==> {result['data']}')
            editeur.linkedin = None

    elif task == 'extract_linkedin_url_from_google_based_on_company_name':
        if i == 0:
            linkedinAutomation.get_google_page()
        result = linkedinAutomation.get_linkedin_url_from_company_name(dict.get('name'))
        if result['status']:
            editeur.linkedin = result['data']
            print(f'{dict.get('name')} ==> {result['data']}')
        else:
            print(f'{dict.get('name')} ==> {result['data']}')
            editeur.linkedin = None

    elif task == 'extract_founder_and_manager_profiles_based_on_keywords':
        result = linkedinAutomation.extract_founder_and_manager_profiles_based_on_keywords(dict)
        if result['status']:
            print(f'{dict.get('name')} ==> {len(result['data'])}')
            for profile in result['data']:
                editeur.profile_name = profile['profile_name']
                editeur.profile_description = profile['profile_description']
                editeur.profile_url = profile['profile_url']
                editeur.profiles = None
                editeurs.append(copy.deepcopy(editeur))
        else:
            print(f'{dict.get('name')} ==> {result['data']}')

    if not editeurs:
        editeurs = editeur
    ItemStorage(file_path='./results/editeurs_de_logiciels_13_have_between_51_and_1000_employees_with_linkedin_about_info_25-11-2024_19h30min.csv', value=editeurs)
    time.sleep(random.uniform(0.5,1))
```

```python
itemStorage = ItemStorage(file_path='./results/editeurs_de_logiciels_13_have_between_51_and_1000_employees_with_linkedin_about_info_25-11-2024_19h30min.csv', value=editeurs)
#itemStorage.data = itemStorage.data.loc[itemStorage.data['company_size'].astype(str) != 'nan']
itemStorage.data = itemStorage.data.sort_values(by='index')
itemStorage.file_path = './results/editeurs_de_logiciels_13_have_between_51_and_1000_employees_with_linkedin_about_info_25-11-2024_19h30min.xlsx'
itemStorage.save_data()
```
